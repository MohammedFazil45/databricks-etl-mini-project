```python
employees_df = spark.table("employees_clean")
dept_avg_df = spark.table("dept_avg_salary")

# Join and filter
above_avg_df = employees_df.join(
    dept_avg_df,
    "Department"
).filter("Salary > avg_salary")

# Save final analytics output
above_avg_df.write.mode("overwrite").saveAsTable("employees_above_dept_avg")
```
