```python
from pyspark.sql.functions import avg, when

df = spark.table("employees_clean")

# Average salary per department
dept_avg_df = df.groupBy("Department") \
    .agg(avg("Salary").alias("avg_salary"))

dept_avg_df.write.mode("overwrite").saveAsTable("dept_avg_salary")

# Add salary category
salary_category_df = df.withColumn(
    "Salary_Category",
    when(df.Salary >= 90000, "High")
    .when(df.Salary >= 50000, "Medium")
    .otherwise("Low")
)

salary_category_df.write.mode("overwrite").saveAsTable("employees_salary_category")
```
