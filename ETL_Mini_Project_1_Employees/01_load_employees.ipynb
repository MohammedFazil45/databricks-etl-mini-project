```python
# initial commit to GitHub


# Load employees table
df = spark.table("employees")

# Basic validation
if df.count() == 0:
    raise Exception("Employees table is empty")

# Optional: remove duplicates
df_clean = df.dropDuplicates(["EmployeeID"])

# Save cleaned data
df_clean.write.mode("overwrite").saveAsTable("employees_clean")
```
