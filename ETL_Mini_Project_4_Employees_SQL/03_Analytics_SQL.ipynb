```sql
--Tranforming and Loading into employees_above_dept_avg Table
MERGE INTO employees_above_dept_avg t
USING(
  SELECT e.*,d.avg_salary 
  from employees_clean e join dept_avg_salary d
  ON e.Department = d.Department AND e.Salary > d.avg_salary
)s
ON t.EmployeeID = s.EmployeeID

WHEN MATCHED AND (
    t.Name <> s.Name OR
    t.Department <> s.Department OR
    t.Role <> s.Role OR
    t.Salary <> s.Salary OR
    t.Experience <> s.Experience OR
    t.Location <> s.Location OR
    t.avg_salary <> s.avg_salary)
THEN
  UPDATE SET
    t.Name = s.Name,
    t.Department = s.Department,
    t.Role = s.Role,
    t.Salary = s.Salary,
    t.Experience = s.Experience,
    t.Location = s.Location = s.Department,
    t.avg_salary = s.avg_salary

WHEN NOT MATCHED THEN
  INSERT (EmployeeID,Name,Department,Role,Salary,Experience,Location,avg_salary)
  VALUES (s.EmployeeID,s.Name,s.Department,s.Role,s.Salary,s.Experience,s.Location,s.avg_salary);
```
